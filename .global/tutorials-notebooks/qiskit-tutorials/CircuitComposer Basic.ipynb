### Simulation and getting the results

Now, it's the right time to see the results!

In here, IBM Quantum Experience ID, or any other providers are temporary not available - we just use the local Qiskit Aer simulator to see the results.

```python
# Display a histogram on 2-D pane
from qiskit import *
from qiskit.visualization import plot_histogram

# Print the result 10 times

i = 0
while i < 100:
    backend = BasicAer.get_backend('qasm_simulator')
    circ = circuit.compose(circuit)
    result = backend.run(transpile(circ, backend), shots=1000).result()
    counts  = result.get_counts(circ)
    print("Counts " + str(i) + ": ", counts)
    i += 1

# Print the last
plot_histogram(counts)
```

```python
# Show the results in Qsphere

from qiskit.visualization import plot_state_qsphere

backend = BasicAer.get_backend('statevector_simulator')
result = backend.run(transpile(circuit, backend)).result()
psi  = result.get_statevector(circuit)
plot_state_qsphere(psi)
```
